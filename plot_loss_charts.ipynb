```python
import numpy as np
import matplotlib.pyplot as plt
import os
import h5py
import pandas as pd
import glob
import cv2
from PIL import Image
```

```python
name = 'Path_Multicamera'
subdir = 'version_1'
file_name = f'{name.lower()}_history.npy'
work_dir = f'/home/adauton/Documents/Mestrado/Tese/Models/{name}/{subdir}'
history = np.load(os.path.join(work_dir, file_name), allow_pickle=True).tolist()
train_loss = np.array(history['loss'])
val_loss = np.array(history['val_loss'])
epochs = len(val_loss)
x_axis = np.arange(epochs)
min_val_loss_index = np.where(val_loss == min(val_loss))[0]
c = [val_loss[min_val_loss_index] if i == min_val_loss_index else None for i in range(epochs)]
```

```python
f = plt.figure()
f.set_figwidth(12)
f.set_figheight(8)

# f.supxlabel('epochs', fontweight ="normal",fontsize=18)

# f.suptitle('this is a nice title', fontweight ="normal",fontsize=18)
patience = 30
final_index = int(min_val_loss_index + patience)

plt.plot(x_axis[:final_index], train_loss[:final_index])
plt.plot(x_axis[:final_index], val_loss[:final_index])
plt.plot(x_axis[:final_index], c[:final_index], 'bo')

plt.legend(['train loss', 'val loss', f'min at epoch {min_val_loss_index + 1}'],prop={'size': 18})

plt.show()
```
