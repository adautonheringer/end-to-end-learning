```python
import numpy as np
import matplotlib.pyplot as plt
import os
import h5py
import glob
import cv2
from PIL import Image
```

```python
DATADIR = '/home/adauton/Documents/Mestrado/Tese/PythonAPI/dataset/'

data = np.load(DATADIR + '/00002822_center.npz')
image = data['x']
label = data['y'][:3].astype(float)
speed = data['y'][4].astype(float)
speed_limit = data['y'][4:6].astype(float)
hlc = np.zeros((4,), dtype='uint8')

command = data['y'][3]
if command == 'Left':
    hlc[0] = 1
elif command == 'Right':
    hlc[1] = 1
elif command == 'Straight':
    hlc[2] = 1
else:
    hlc[3] = 1

print(image.dtype)
print(label)
print(data['y'][3])
print(hlc)
print(speed)
print(speed_limit)
print(label[2] == 0)
plt.imshow(image)
plt.show()
```

```python
X = []
Y = []
speed_and_limit = []
HLC = []

for file_name in os.listdir(DATADIR):
    data = np.load(DATADIR + file_name)
    x = data['x']
    y = data['y'][:3]
    _speed_and_limit = data['y'][4:6]
    hlc = np.zeros((4,), dtype='uint8')

    command = data['y'][3]
    
    if command == 'Left':
        hlc[0] = 1
    elif command == 'Right':
        hlc[1] = 1
    elif command == 'Straight':
        hlc[2] = 1
    else:
        hlc[3] = 1
    
    X.append(x)
    Y.append(y)
    speed_and_limit.append(_speed_and_limit)
    HLC.append(hlc)
```

```python
npX = np.array(X)
npHlc = np.array(HLC)
npY = np.array(Y)
```

```python
print(npHlc)
```

```python


np.savez('dataset/20_rounds', IMAGES=npX, Y=npY, HLC=npHlc)
```
